### #10872 팩토리얼
* https://www.acmicpc.net/problem/10872

```python
def factorial(n) :
    if n == 1 or n==0:
        return 1
    return n*factorial(n-1)

print(factorial(int(input())))
```

### #10870 피보나치 수 5
* https://www.acmicpc.net/problem/10870

```python
def fibo(n) :
    if n == 0 :
        return 0
    if n == 1 :
        return 1
    return fibo(n-1) + fibo(n-2)

print(fibo(int(input())))
```

### #25501 재귀의 귀재
* stdin.readline().rstrip()
* https://mgyo.tistory.com/166
* https://www.acmicpc.net/problem/25501 

```python
import sys
input = sys.stdin.readline

def recursion(s, l, r, n) :
    if l >= r :
        return 1, n
    elif s[l] != s[r] :
        return 0, n
    else :
        n += 1
        return recursion(s, l+1, r-1, n)

def isPalindrome(s) :
    return recursion(s, 0, len(s)-1, 1)

t = int(input()) #int로 변환시에는 rstrip을 붙이지 않아도 개행문자(\n)가 제거됨
for i in range(t) :
    data = isPalindrome(input().rstrip())
    print(data[0], data[1])
```

### #24060 알고리즘 수업 - 병합 정렬 1
* 다소 비효율적인 코드
* https://www.acmicpc.net/problem/24060

```python
n, k = map(int,input().split()) #배열의 크기 #저장횟수
li = list(map(int,input().split())) #배열의 원소

savenum = 0
save = 0

def merge_sort(data, p, r, k) :
    if p < r :
        q = (p+r)//2
        merge_sort(data, p, q, k) #전반부 정렬
        merge_sort(data, q+1, r, k) #후반부 정렬
        return merge(data, p, q, r, k) #병합


def merge(data, p, q, r, k) :
    global savenum, save
    i = p #전반부 시작지점
    j = q+1 #후반부 시작지점
    t = 0 #tmp 리스트의 포인터
    tmp = []
    while (i <= q and j <= r) :
        if data[i] <= data[j] :
            tmp.append(data[i])
            i += 1 
        else :
            tmp.append(data[j])
            j += 1
    while i <= q : #전반부 배열 부분이 남은 경우
        tmp.append(data[i])
        i += 1
    while j <= r : #후반부 배열 부분이 남은 경우
        tmp.append(data[j])
        j += 1
    i = p
    t = 0
    while i <= r : #결과를 data에 저장
        data[i] = tmp[t]
        savenum += 1
        if savenum == k :
            save = tmp[t]
        i += 1
        t += 1
    return data

merge_sort(li, 0, len(li)-1, k)

if savenum < k :
    print('-1')
else :
    print(save)
```

### #2447 별 찍기 - 10
* https://www.acmicpc.net/problem/2447

```python
import math

def star(n):
    s = []
    if n==3 :
        s.append('***')
        s.append('* *')
        s.append('***')
    elif n>=9 :
        t = star(n//3)
        for i in range(len(t)):
            s.append(t[i]*3)
        for i in range(len(t)):
            s.append(t[i] + ' '*(n//3) + t[i])
        for i in range(len(t)):
            s.append(t[i]*3)
    return s
n = int(input())
s = star(n)
for i in range(len(s)) :
    print(s[i])
```

### #11729 하노이 탑 이동 순서
* https://shoark7.github.io/programming/algorithm/tower-of-hanoi
* **재귀함수의 탈출 조건, 또는 기저 사례(base case) : 옮기는 원판이 1개일 때**
* https://www.acmicpc.net/problem/11729

```python
k = int(input())

def move(start, to) :
    print(start, to)
    return

def hanoi(k, start, via, to) :
    if k == 1 :
        move(start, to)
        return
    else :
        hanoi(k-1,start, to, via) #제일 큰 원판 위의 원판들을 초기의 경유지점으로 옮김
        move(start, to) #초기의 목적지로 가장 큰 원판을 옮김 
        hanoi(k-1, via, start, to) #초기의 경유지점에서 초기의 목적지로 옮김 (이때의 경유지점은 시작지점)
    return
print(2**k-1)
hanoi(k,1,2,3)
```
